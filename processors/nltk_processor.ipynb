```python
import requests
import time
import pandas as pd

SERVER_URL = 'http://0.0.0.0:6543/'
```

```python
# Combine datasets 
df1 = pd.read_csv('../clean_data/res_gutenberg.csv', encoding = "ISO-8859-1", names=["word", "meaning"])
df2 = pd.read_csv('../clean_data/res_wordmeaning.csv', names=["word", "meaning"])
df3 = pd.read_csv('../clean_data/res_wordnet20.csv', names=["word", "meaning"])
```

```python
df = pd.concat([df1, df2, df3])
```

```python
df.describe()
```

```python
def get_nltk_meanings(word):
    query = '?q={}'.format(word)
    url = SERVER_URL + query
    res = requests.get(url)
    return res.json()
```

```python
res = []

words = sorted(df.word.dropna().unique())

print(words)
total = len(words)
for idx, word in enumerate(words):
    try:
        if (idx % 1000) == 0:
            print(idx, 'of', total)
            time.sleep(1)
        meanings = get_nltk_meanings(word)
        if meanings:
            for m in meanings[word]:
                res.append((word, m))
    except:
        print(word)
        continue
```

```python
len(res)
```

```python
import csv
with open('res_nltk.csv', 'w') as outfile:
    writer = csv.writer(outfile)
    for row in res:
        writer.writerow(row)
```
